```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

plt.style.use('ggplot')
print("✅ Machine Learning Stack listo")
```

## 1. Segmentación de Flota (Clustering)
Buscamos agrupar viajes con perfiles de eficiencia similares.

```python
# Generamos datos de flota ficticios
np.random.seed(42)
viajes = pd.DataFrame({
    'distancia_nm': np.random.uniform(500, 5000, 500),
    'carga_ton': np.random.uniform(10000, 50000, 500)
})
viajes['consumo_ton_dia'] = (viajes['carga_ton'] * 0.001) + (viajes['distancia_nm'] * 0.01) + np.random.normal(0, 2, 500)

# Escalado y Entrenamiento
X = StandardScaler().fit_transform(viajes)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
viajes['cluster'] = kmeans.labels_

# Visualización
plt.scatter(viajes['distancia_nm'], viajes['consumo_ton_dia'], c=viajes['cluster'], cmap='viridis', alpha=0.6)
plt.title("Segmentación de Viajes por Eficiencia")
plt.xlabel("Distancia (NM)")
plt.ylabel("Consumo (Ton/Día)")
plt.show()
```

## 2. Detección de Anomalías (Fraud & Error Detection)
Uso de Isolation Forest para detectar consumos de combustible imposibles o fugas.

```python
iso = IsolationForest(contamination=0.03, random_state=42)
viajes['anomalia'] = iso.fit_predict(viajes[['consumo_ton_dia']])

anomalos = viajes[viajes['anomalia'] == -1]
print(f"⚠️ Se detectaron {len(anomalos)} viajes con consumos anómalos.")
anomalos.head()
```
