```python
import pandas as pd
import numpy as np
import plotly.express as px

print("✅ Visualización Interactiva lista")
```

## 1. Cálculo de Indicadores CII
Implementamos la métrica oficial de intensidad de carbono.

```python
# Datos sintéticos de flota 2023
np.random.seed(8)
flota = pd.DataFrame({
    'buque': [f'Ship-{i}' for i in range(1, 51)],
    'distancia_nm': np.random.uniform(5000, 20000, 50),
    'consumo_hfo_ton': np.random.uniform(300, 1500, 50),
    'dwt': np.random.choice([10000, 25000, 50000, 100000], 50)
})

# Cálculo CII
flota['cii'] = (flota['consumo_hfo_ton'] * 3.114 * 1e6) / (flota['dwt'] * flota['distancia_nm'])

# Calificación simplificada
def rate(val):
    if val < 4: return 'A'
    if val < 6: return 'B'
    if val < 9: return 'C'
    return 'D'

flota['rating'] = flota['cii'].apply(rate)
flota.head()
```

## 2. Dashboard Interactivo con Plotly
Permite inspeccionar buques específicos y sus métricas.

```python
fig = px.scatter(flota, 
                 x="distancia_nm", 
                 y="cii", 
                 color="rating",
                 size="dwt", 
                 hover_name="buque",
                 color_discrete_map={'A':'green', 'B':'blue', 'C':'orange', 'D':'red'},
                 title="Análisis de Intensidad de Carbono (CII) — Flota UMIP 2024")

fig.show()
```
