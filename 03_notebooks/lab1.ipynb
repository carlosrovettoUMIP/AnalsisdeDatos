```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set_theme(style="whitegrid", palette="viridis")
plt.rcParams['figure.figsize'] = (12, 6)

print("✅ Librerías cargadas")
```

## 1. Carga y Preparación de Datos
Intentamos cargar los datos del puerto o generamos una muestra sintética robusta.

```python
def cargar_datos():
    try:
        # Intentar cargar desde GitHub (Colab) o Local
        url = "https://jacenteno.github.io/Analisis_Datos_Maritimo/02_Datasets/atraques_balboa.csv"
        try:
            df = pd.read_csv(url)
            print("✅ Datos cargados desde GitHub")
        except:
            df = pd.read_csv("../02_Datasets/atraques_balboa.csv")
            print("🏠 Datos cargados localmente")
        print("📂 Cargando desde carpeta de Datasets...")
    except:
        print("⚠️ Dataset no encontrado. Generando datos sintéticos para la práctica...")
        np.random.seed(42)
        tipos = ['Portacontenedores', 'Granelero', 'Tanquero', 'RoRo', 'Carga General']
        df = pd.DataFrame({
            'fecha': pd.date_range(start='2018-01-01', periods=1200, freq='D'),
            'nombre_buque': [f'MV-{np.random.randint(1000, 9999)}' for _ in range(1200)],
            'tipo_buque': np.random.choice(tipos, 1200),
            'teu_cargados': np.random.randint(100, 5000, 1200),
            'tiempo_estadia_h': np.random.uniform(12, 72, 1200),
            'costo_usd': np.random.uniform(5000, 25000, 1200)
        })
        df['año'] = df['fecha'].dt.year
    return df

df = cargar_datos()
df.head()
```

## 2. El Reporte (Enfoque Tradicional)
Responde a: *¿Qué pasó?*

```python
resumen_anual = df.groupby('año')['teu_cargados'].sum()
print("📊 TOTAL DE TEUs POR AÑO (Reporte):")
print(resumen_anual)
```

## 3. El Análisis (Enfoque Maestría)
Responde a: *¿Por qué pasó? ¿Qué implicaciones tiene?*

```python
# Análisis de Eficiencia: Costo por TEU vs Tiempo de Estadía
df['costo_por_teu'] = df['costo_usd'] / df['teu_cargados']

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.barplot(data=df, x='tipo_buque', y='costo_por_teu', ax=axes[0])
axes[0].set_title("Costo Medio por TEU movilizado")

sns.scatterplot(data=df, x='tiempo_estadia_h', y='costo_usd', hue='tipo_buque', alpha=0.5, ax=axes[1])
axes[1].set_title("Relación Tiempo de Estadía vs Costo Total")

plt.show()
```
