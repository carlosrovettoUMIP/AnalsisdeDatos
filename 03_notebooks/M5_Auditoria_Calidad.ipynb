## 1. Función de Auditoría Automatizada
Definimos una lógica que evalúa la completitud y consistencia al instante.

```python
def auditoria_dataset(df, nombre="Dataset Marítimo"):
    print(f"\n{'='*60}")
    print(f" 🛡️  AUDITORÍA DE CALIDAD: {nombre}")
    print(f"{'='*60}")
    
    # 1. Dimensiones
    print(f" 📈 Filas: {len(df):,}")
    print(f" 📉 Columnas: {df.shape[1]}")
    
    # 2. Completitud
    print("\n 🧩 COMPLETITUD por Columna:")
    null_pct = (1 - df.isnull().mean()) * 100
    for col, pct in null_pct.items():
        status = "✅ OK" if pct == 100 else "⚠️ Bajo" if pct > 90 else "❌ Crítico"
        print(f"    [{status}] {col:<20} | {pct:>6.1f}%")
    
    # 3. Datos Duplicados
    dups = df.duplicated().sum()
    print(f"\n 🔄 Registros Duplicados: {dups} ({'✅ Consistente' if dups == 0 else '⚠️ Revisar'})")
    
    return null_pct

# Simulando datos de Lloyd's Register
df_lloyd = pd.DataFrame({
    'mmsi': [1, 2, 3, 4, 3, 5],
    'bandera': ['Panamá', 'Liberia', 'Malta', np.nan, 'Malta', 'Panamá'],
    'año_construccion': [2020, 2018, 1995, 2022, 1995, np.nan]
})

res = auditoria_dataset(df_lloyd, "Muestra Lloyd's Register 2024")
```
