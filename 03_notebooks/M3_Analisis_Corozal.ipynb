```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

sns.set_theme(style="white")
print("✅ Librerías listas")
```

## 1. Análisis Descriptivo y Diagnóstico
¿Qué pasó y por qué? Usamos correlaciones para entender cuellos de botella.

```python
# Generamos datos para Corozal 2020-2024
np.random.seed(7)
data = {
    'mes': np.tile(np.arange(1, 13), 5),
    'año': np.repeat(np.arange(2020, 2025), 12),
    'teu_cargados': np.random.randint(5000, 15000, 60) + np.linspace(0, 5000, 60),
    'tiempo_espera_h': np.random.uniform(2, 12, 60),
    'buques_en_puerto': np.random.randint(1, 10, 60),
    'clima_codigo': np.random.randint(0, 3, 60)
}
df = pd.DataFrame(data)

# Matriz de Correlación (Diagnóstico)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn', center=0)
plt.title("Diagnóstico: Correlación de variables operativas")
plt.show()
```

## 2. Análisis Predictivo (Regresión)
¿Qué pasará? Entrenamos un modelo para predecir la demanda de TEUs.

```python
# Preparación de datos
X = df[['mes', 'buques_en_puerto']]
y = df['teu_cargados']

modelo = LinearRegression().fit(X, y)
df['prediccion'] = modelo.predict(X)

# Visualización de la tendencia predictiva
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['teu_cargados'], label='Real', marker='o')
plt.plot(df.index, df['prediccion'], label='Predicción', linestyle='--', color='red')
plt.title("Pronóstico de Demanda de Contenedores — Terminal Corozal")
plt.legend()
plt.show()

print(f"R² del modelo: {modelo.score(X, y):.3f}")
```
